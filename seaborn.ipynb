# Bar plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
sn.get_dataset_names()
```

```python
tips=sn.load_dataset('tips')
```

```python
tips.head()
```

Draw a simple Bar Plot


```python
sn.barplot(x='day',y='total_bill',data=tips)
```

Hue Attribute

```python
sn.barplot(x='day',y='total_bill',hue='sex',data=tips)
```

```python
sn.barplot(x='day',y='total_bill',ci=60,capsize=0.3,data=tips)
```

Saturation level

```python
sn.barplot(x='day',y='total_bill',color='green',saturation=0.1,data=tips)
```

# Displots (Distribution Plots)

```python
pen = sn.load_dataset('penguins')
```

```python
pen.head()
```

Draw a Simple displot


```python
sn.distplot(pen.flipper_length_mm,rug=True)
```

Removing KDE plot

```python
sn.distplot(pen.flipper_length_mm,hist=False)
```

Number of bins

```python
sn.distplot(pen.flipper_length_mm,bins=20)
```

Setting bin values as array

```python
sn.distplot(pen.flipper_length_mm,bins=[20,160,180,190,200,230])
```

Fitting the data to a particular distribution using norm_hist

```python
sn.distplot(pen.flipper_length_mm,norm_hist=True,kde=False)
```

Styling Rugplots

```python
sn.distplot(pen.flipper_length_mm,rug_kws={'color':'red','height':0.30},rug=True)
```

Histogram Styling

```python
sn.distplot(pen.flipper_length_mm,hist_kws={'color':'blue','alpha':0.3,'histtype':'step'})#'histtype' :'stepfilled'
```

styling KDE plots

```python
sn.distplot(pen.flipper_length_mm,kde=True,hist=False,kde_kws={'color':'red','lw':2.5,'alpha':0.4})
```

# Box plots in seaborn


*   Known as Categorical Distribution plots




```python
pen.describe()
```

Basic Boxplot

```python
sn.boxplot(pen.bill_length_mm)
```

```python
sn.boxplot(y=pen.bill_length_mm,x=pen.island)
```

Hue Attribute

```python
sn.boxplot(x=pen.bill_length_mm,y=pen.island,hue=pen.sex)
```

ordering of Bars

```python
sn.boxplot(x=pen.bill_length_mm,y=pen.island,hue=pen.sex,order=['Biscoe','Torgersen','Dream'])
```

color Attribute

```python
sn.boxplot(y=pen.bill_length_mm,x=pen.island,hue=pen.sex,order=['Biscoe','Torgersen','Dream'],color='y')
```

width property

```python
sn.boxplot(x=pen.bill_length_mm,y=pen.island,hue=pen.sex,order=['Biscoe','Torgersen','Dream'],width=0.5)
```

Whis property

```python
sn.boxplot(x=pen.bill_length_mm,y=pen.island,hue=pen.sex,order=['Biscoe','Torgersen','Dream'],whis=0.5)
```

showcase property

```python
sn.boxplot(x=pen.bill_length_mm,y=pen.island,hue=pen.sex,order=['Biscoe','Torgersen','Dream'],showcaps=False)
```
