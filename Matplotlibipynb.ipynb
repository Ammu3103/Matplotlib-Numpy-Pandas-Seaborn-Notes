# Intro to matplotlib


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,44,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='red',linewidth=4, linestyle="dotted")
```

# Format String


```python
plt.plot(x,y,'g+')
```

```python
plt.plot(x,y,'g+--')
```

```python
plt.plot(x,y,'g+-.')
```

```python
plt.plot(x,y,'--r*')
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='')
```

```python
plt.plot(x,y,color='red',marker='H',linestyle='',markersize=20)
```

#alpha -- Transparency


```python
#alpha -- Transparency
plt.plot(x,y,color='green',marker='^',linestyle='',markersize=20,alpha=0.2)
```

# Axes labels, Legend, Grid


```python
Days =[1,2,3,4,5,6,7]
max_t = [52,54,50,44,49,51,46]
min_t=[50,52,48,42,47,49,44]
avg_t = [51,53,49,43,48,50,45]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('weather')
plt.plot(Days,max_t,label='Max')
plt.plot(Days,min_t,label='Min')
plt.plot(Days,avg_t,label='Avg')

plt.legend(loc='best',shadow='True',fontsize='large')

plt.grid()
```

# Bar Chart


```python
import numpy as np
```

```python
company =['Google','Amazon','Microsoft','Facebook']
Revenue = [90,120,87,30]
profit= [40,20,30,15]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(bln)')
plt.title('US Tech stocks')
plt.bar(xpos-0.2,Revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2,profit, width=0.4,label='Profit')
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.title('US Tech stocks')

plt.barh(xpos-0.2,Revenue, label='Revenue')
plt.barh(xpos+0.2,profit,label='Profit')
plt.legend()
```

# Histograms


```python
blood_sugar = [113,180,50,60,70,120,100,115,135,80,129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g',alpha=0.5)#histtype='step'
```

```python
blood_sugar_men = [113,180,50,60,70,120,100,115,135,80,129]
blood_sugar_women = [100,150,100,160,97,102,70,105,35,170,85]
plt.xlabel('Sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],alpha=0.5,label=['Men','Women'])#histtype='step,orientation='horizontal'
plt.legend()
```

# Pie chart


```python
expenditure = [6000,1500,500,1000,200]
exp_items =['Home rent','Grocery','Phone/Internet bill','Car','Other utillities']
plt.pie(expenditure,labels=exp_items,radius=1.25, autopct='%0.1f%%',shadow=True,explode=[0,0.1,0,0,0],startangle=45)
```

# Save chart

```python
expenditure = [6000,1500,500,1000,200]
exp_items =['Home rent','Grocery','Phone/Internet bill','Car','Other utillities']
plt.pie(expenditure,labels=exp_items,autopct='%0.1f%%')
plt.savefig("pie.png",bbox_inches="tight",pad_inches=1,transparent=True)
```
